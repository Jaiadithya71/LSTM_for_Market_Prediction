```python
%%writefile /content/app.py
import streamlit as st
import pandas as pd
import numpy as np
import tensorflow as tf
from stock_predictor import train_stock_model
from PIL import Image
import os
from datetime import datetime, timedelta
import time
from TradingView_Data.TradingviewData.main import TradingViewData, Interval
from dataset_fetcher import fetch_datasets_for_indices, save_datasets # Assumed module for fetching historical and live data
import dataset_fetcher
import pytz

# Streamlit app configuration
st.set_page_config(page_title="Stock Price Prediction Dashboard", layout="wide")
st.title("Stock Price Prediction Dashboard")

# Indices to process with NSE_NIFTY as default
indices_to_process = [
    "NSE_NIFTY",  # Prioritized as default
    "TVC_DJI",
    "TVC_SPX",
    "NASDAQ_DLY_NDX",
    "TVC_NI225",
    "SSE_DLY_000001",
    "ATHEX_DLY_FTSE",
    "TVC_DXY",
    "EASYMARKETS_DAXEUR"
]

# Sidebar for index selection
st.sidebar.header("Select Index")
selected_index = st.sidebar.selectbox("Choose an index:", indices_to_process, index=0)

# Fetch historical data using dataset_fetcher
try:
    datasets = fetch_datasets_for_indices(indices_to_process, 8000)  # Fetch all datasets
    historical_data = datasets[selected_index]  # Extract DataFrame for selected_index
    historical_data = historical_data.tail(6000)  # Ensure manageable size (use tail instead of slice for consistency)
except Exception as e:
    st.error(f"Error fetching historical data for {selected_index}: {str(e)}")
    st.stop()

# Validate historical data
required_columns = ['time', 'open', 'high', 'low', 'close']
if not all(col in historical_data.columns for col in required_columns):
    st.error(f"Historical data must contain columns: {required_columns}")
    st.stop()

# Load predictions from CSV (if model has been trained)
predictions_file = f"output/{selected_index}_15_15min/{selected_index}_15_15min_predictions.csv"
try:
    predictions_df = pd.read_csv(predictions_file)
except FileNotFoundError:
    st.warning(f"Predictions for {selected_index} not found. Please train the model by clicking 'Live Price Prediction'.")
    predictions_df = None

# Display last 10 records with predictions
st.subheader(f"Last 10 Records for {selected_index}")
if predictions_df is not None:
    # Merge historical data with predictions
    historical_data['time'] = pd.to_datetime(historical_data['time'])
    predictions_df['datetime'] = pd.to_datetime(predictions_df['datetime'])
    merged_df = historical_data.merge(predictions_df, left_on='time', right_on='datetime', how='inner')
    last_10 = merged_df.tail(10)[['time', 'open', 'high', 'low', 'close', 'predicted', 'actual']]
    last_10.columns = ['Time', 'Open', 'High', 'Low', 'Close', 'Predicted next Close', 'Actual next Close']
    st.dataframe(last_10)
else:
    last_10 = historical_data.tail(10)[required_columns]
    last_10.columns = ['Time', 'Open', 'High', 'Low', 'Close']
    st.dataframe(last_10)

# Display accuracy metrics and plot
plot_file = f"output/{selected_index}_15_15min/{selected_index}_15_15min_prediction_plot.png"
features_file = f"output/{selected_index}_15_15min/Features.csv"
model_file = f"output/{selected_index}_15_15min/{selected_index}_15_15min_prediction_model.keras"

if os.path.exists(features_file):
    features_df = pd.read_csv(features_file)
    st.subheader("Model Features")
    st.write(features_df)

if os.path.exists(plot_file):
    st.subheader("Prediction Plot")
    plot_image = Image.open(plot_file)
    st.image(plot_image, caption="Actual vs Predicted Close Prices")

# Live prediction state
if 'live_mode' not in st.session_state:
    st.session_state.live_mode = False
if 'last_candle' not in st.session_state:
    st.session_state.last_candle = None
if 'next_close_pred' not in st.session_state:
    st.session_state.next_close_pred = None
if 'last_update_time' not in st.session_state:
    st.session_state.last_update_time = datetime.now()

# Live Price Prediction Button
if st.button("Live Price Prediction"):
    st.session_state.live_mode = True
    # Fetch live data using dataset_fetcher
    try:
        live_data = dataset_fetcher.fetch_dataset(
            symbol_key=selected_index,
            n_bars=6000  # Fetch latest data
        )
        if live_data is None:
            raise ValueError("No live data returned from dataset_fetcher.")
        live_data = live_data[required_columns]
    except Exception as e:
        st.error(f"Error fetching live data: {str(e)}")
        st.session_state.live_mode = False
        st.stop()

    # Combine historical and live data
    combined_data = pd.concat([historical_data, live_data]).drop_duplicates(subset='time').sort_values('time')

    # Retrain model
        # Retrain model with simulated progress bar
    try:
        total_epochs = 20
        progress_bar = st.progress(0)
        status_text = st.empty()

        # Estimate time per epoch (e.g., assume 2 seconds per epoch, adjust based on actual performance)
        time_per_epoch = 2  # Adjust this value based on your system's performance
        total_time = total_epochs * time_per_epoch

        # Start training
        import threading
        import time

        # Function to update progress bar in a separate thread
        def update_progress():
            for epoch in range(total_epochs):
                time.sleep(time_per_epoch)
                progress = (epoch + 1) / total_epochs
                progress_bar.progress(min(progress, 1.0))
                status_text.text(f"Training model: {int(progress * 100)}% (Epoch {epoch + 1}/{total_epochs})")

        # Run progress update in a separate thread
        progress_thread = threading.Thread(target=update_progress)
        progress_thread.start()

        # Train the model
        results = train_stock_model(
            dataset=combined_data,
            seq_length=60,
            train_split=0.7,
            epochs=20,
            batch_size=32,
            model_filename=f"{selected_index}_15_15min.csv",
            output_dir="output"
        )
        st.session_state.results = results

        # Wait for the progress thread to finish (if training finishes faster)
        progress_thread.join()
        progress_bar.progress(1.0)  # Ensure progress bar reaches 100%
        status_text.empty()  # Clear the status text
        st.success("Model retrained with live data!")
    except Exception as e:
        st.error(f"Error retraining model: {str(e)}")
        st.session_state.live_mode = False
        st.stop()

# Function to compute EMAs
def compute_emas(df, periods=[5, 10, 15]):
    """
    Compute Exponential Moving Averages (EMAs) for the given periods and add them to the DataFrame.

    Args:
        df (pandas.DataFrame): DataFrame with 'close' column
        periods (list): List of periods for which to compute EMAs (e.g., [5, 10, 15])

    Returns:
        pandas.DataFrame: DataFrame with additional EMA columns
    """
    df = df.copy()
    for period in periods:
        df[f'ema_{period}'] = df['close'].ewm(span=period, adjust=False).mean()
    return df

# Function to predict next close price
def predict_next_close(model, feature_scaler, target_scaler, data, features, seq_length):
    # Prepare the latest sequence
    data = compute_emas(data)
    latest_data = data.tail(seq_length)[features]
    scaled_data = feature_scaler.transform(latest_data)
    sequence = np.array([scaled_data])

    # Predict
    pred_scaled = model.predict(sequence)
    pred_actual = target_scaler.inverse_transform(pred_scaled)
    return pred_actual[0][0]

# Live mode: Fetch last traded price and predict next close
if st.session_state.live_mode:
    # Fetch the latest candlestick using dataset_fetcher
    try:
        latest_data = dataset_fetcher.fetch_live_data(
            selected_index
        )
        latest_data = latest_data[required_columns]
        st.session_state.last_candle = latest_data.iloc[-1]
        st.session_state.last_update_time = datetime.now()
    except Exception as e:
        st.error(f"Error fetching latest candlestick: {str(e)}")
        st.stop()

    # Predict next close
    if os.path.exists(model_file):
        from tensorflow.keras.models import load_model
        model = load_model(model_file)
        results = st.session_state.results
        next_close = predict_next_close(
            model,
            results['feature_scaler'],
            results['target_scaler'],
            pd.concat([historical_data, latest_data]).sort_values('time'),
            ['open', 'high', 'low', 'close', 'ema_5', 'ema_10', 'ema_15'],
            seq_length=60
        )
        st.session_state.next_close_pred = next_close

    # Display last traded price and predicted close
    st.subheader("Last Traded Price and Next Predicted Close")
    if st.session_state.last_candle is not None:
        last_candle = st.session_state.last_candle
        last_candle_df = pd.DataFrame({
            'Time': [last_candle['time']],
            'Open': [last_candle['open']],
            'High': [last_candle['high']],
            'Low': [last_candle['low']],
            'Close': [last_candle['close']],
            'Predicted Next Close': [st.session_state.next_close_pred]
        })
        # Highlight Predicted Next Close
        def highlight_pred(s):
            return ['background-color: #333333' if s.name == 'Predicted Next Close' else '' for _ in s]
        st.dataframe(last_candle_df.style.apply(highlight_pred, axis=0))


    last_candle_time = pd.to_datetime(st.session_state.last_candle['time'])
    next_candle_time = last_candle_time + timedelta(minutes=15)
    kolkata_tz = pytz.timezone("Asia/Kolkata")
    current_time = datetime.now(kolkata_tz)

    if next_candle_time.tzinfo is None:
        # Assume it's in Asia/Kolkata if not specified
        next_candle_time = kolkata_tz.localize(next_candle_time)
    time_to_next = (next_candle_time - datetime.now(kolkata_tz)).total_seconds()

    st.subheader("Next Candlestick Timer")
    while (time_to_next > 0):
        st.write(f"Next candlestick in: {int(time_to_next)} seconds")
        # Simulate timer countdown
        placeholder = st.empty()
        while time_to_next > 0:
            time_to_next = (next_candle_time - datetime.now(kolkata_tz)).total_seconds()
            placeholder.write(f"Next candlestick in: {int(time_to_next)} seconds")
            time.sleep(1)

    # Fetch new candlestick
    new_data = dataset_fetcher.fetch_live_data(
        symbol_key=selected_index,  # Fixed missing parameter name
        n_bars=1
    )

    if new_data is None:
        st.error("Failed to fetch new candlestick.")
        st.stop()
    st.write("Fetched new data:")
    st.write(new_data)
    new_candle = new_data.iloc[-1]

    # Calculate errors only if predicted_close is available
    if st.session_state.next_close_pred is not None:
        actual_close = new_candle['close']
        predicted_close = st.session_state.next_close_pred
        abs_error = abs(actual_close - predicted_close)
        if actual_close != 0:
            percent_error = (abs_error / actual_close) * 100
        else:
            percent_error = float('inf')

        st.subheader("Prediction Error")
        st.write(f"Actual Close: {actual_close:.2f}")
        st.write(f"Predicted Close: {predicted_close:.2f}")
        st.write(f"Absolute Error: {abs_error:.2f}")
        st.write(f"Percentage Error: {percent_error:.4f} %")
    else:
        st.warning("No prediction available. Please ensure the model is trained by clicking 'Live Price Prediction'.")

        # Update last candle and predict again
        st.session_state.last_candle = new_candle
        st.session_state.last_update_time = datetime.now()
        next_close = predict_next_close(
            model,
            results['feature_scaler'],
            results['target_scaler'],
            pd.concat([historical_data, new_data]).sort_values('time'),
            ['open', 'high', 'low', 'close', 'ema_5', 'ema_10', 'ema_15'],
            seq_length=60
        )
        st.session_state.next_close_pred = next_close
        st.experimental_rerun()

# Download button for last 100 records
st.subheader("Download Data")
if predictions_df is not None:
    last_100 = merged_df.tail(100)[['time', 'open', 'high', 'low', 'close', 'predicted']]
    last_100.columns = ['Time', 'Open', 'High', 'Low', 'Close', 'Predicted Close']
    csv = last_100.to_csv(index=False)
    st.download_button(
        label="Download Last 100 Records",
        data=csv,
        file_name=f"{selected_index}_last_100_records.csv",
        mime="text/csv"
    )
else:
    st.write("No predictions available to download. Please train the model by clicking 'Live Price Prediction'.")

# Sidebar instructions
st.sidebar.header("Instructions")
st.sidebar.write("1. Select an index from the dropdown.")
st.sidebar.write("2. View the last 10 records, predictions, plot, and metrics.")
st.sidebar.write("3. Click 'Live Price Prediction' to fetch live data and retrain the model.")
st.sidebar.write("4. View the last traded price, predicted next close, and timer.")
st.sidebar.write("5. When the timer hits zero, errors are calculated, and the process repeats.")
st.sidebar.write("6. Download the last 100 records using the button.")
```
